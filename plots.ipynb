### Performance analysis

```python
import matplotlib.pyplot as plt
import numpy as np

opacity = 0.5
bar_width = 0.35

# [Decision Tree, Random Forest, KNN, MLP, Nearest Centroids]
data_114_genres = [0.3, 0.5, 0.35, 0.4, 0.14]
data_20_genres = [0.4, 0.65, 0.4, 0.55, 0.28]


x_coords = np.arange(len(data_114_genres))
bar_113_genres = plt.bar(x_coords + bar_width, data_114_genres, bar_width,
               align='center', alpha=opacity, color='darkorange', label='114 genres')
bar_20_genres = plt.bar(x_coords, data_20_genres, bar_width,
               align='center', alpha=opacity, color='g', label='20 genres')

# Add the sports for each bar and the height
for bar in bar_113_genres + bar_20_genres:
    height = bar.get_height()
    # Place height of each bar text at the top of the bar
    plt.text(bar.get_x() + bar.get_width()/2.0, height, f'{height:.2f}',
             ha='center', va='bottom')

plt.legend()
plt.ylabel('F1-score')
plt.title('Performance analysis for different ML methods')
plt.xticks(x_coords + bar_width/2.0,('Decision Tree', 'Random Forest', 'KNN', 'MLP', 'Nearest Centroids'))
plt.tight_layout()
plt.show()
```
